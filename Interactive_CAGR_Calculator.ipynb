**STEP 1**: Import all necessary libraries



```python
import math
from ipywidgets import widgets, VBox, HBox, Button, Output
from IPython.display import display, clear_output
```

**STEP 2**: Computation of the CAGR and the design of the UI for the calculator

> A - Ensuring that the initial investment, closing investment and the years value are not less than 0. Defining the CAGR formula

> B - Designing the UI. Creating the initial investment value label and input box, closing investment value label and input box, year label and input box, and a space for the result of the CAGR computation (Output)

> C - Ensuring that the commas do not negate any input value or the output

> D - Displaying the UI

```python
def calculate_cagr(initial_investment_value, current_investment_value, years):
    try:
        if initial_investment_value <= 0 or current_investment_value <= 0 or years <= 0:
            return "All inputs must be positive numbers greater than 0."

        cagr = ((current_investment_value / initial_investment_value) ** (1 / years)) - 1
        return f"{cagr * 100:.2f}%"
    except Exception as e:
        return f"Error in calculation: {str(e)}"


initial_investment_value_label = widgets.Label(value="Initial Investment Value (Naira):")
initial_investment_value_input = widgets.Text(placeholder="Enter start value", style={'description_width': 'initial'})

current_investment_value_label = widgets.Label(value="Current Investment Value (Naira):")
current_investment_value_input = widgets.Text(placeholder="Enter end value", style={'description_width': 'initial'})

years_label = widgets.Label(value="Years:")
years_input = widgets.Text(placeholder="Enter number of years", style={'description_width': 'initial'})

output = Output()

calc_button = Button(description="Calculate CAGR", button_style='success')

def on_button_click(b):
    clear_output(wait=True)
    output.clear_output()
    with output:
        try:
            initial_investment_value = float(initial_investment_value_input.value.replace(",", ""))
            current_investment_value = float(current_investment_value_input.value.replace(",", ""))
            years = float(years_input.value)

            result = calculate_cagr(initial_investment_value, current_investment_value, years)
            print(f"The compounding annual growth rate for your investment is {result}")
        except ValueError:
            print("Dear user, the details you entered are not numerical. Kindly take a look at your inputs again and click the 'Calculate CAGR' button")

calc_button.on_click(on_button_click)

ui = VBox([
    HBox([initial_investment_value_label, initial_investment_value_input],layout=widgets.Layout(margin="15px 0")),
    HBox([current_investment_value_label, current_investment_value_input], layout=widgets.Layout(margin="15px 0")),
    HBox([years_label, years_input], layout=widgets.Layout(margin="15px 0")),
    HBox([calc_button], layout=widgets.Layout(margin="15px 0")),
    HBox([output], layout=widgets.Layout(margin="15px 0"))
])

display(ui)
```
